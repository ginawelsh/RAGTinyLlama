```python
## compare responses against original text

import torch
import csv
from collections import defaultdict
import pandas as pd
from bert_score import score
from sklearn.feature_extraction.text import TfidfVectorizer

def process_shot_files(path):
  topic_text_dict = defaultdict(list)
  with open(path, "r") as file:
    next(file)
    for row in csv.reader(file):
      text = row[3]
      topic = row[1].lower().replace("_", " ")
      topic_text_dict[topic].append(text)
  return topic_text_dict

def get_relevant_text(question, source_text, n_sentences=3):
    sentences = [s.strip() for s in source_text.split('.') if s.strip()]
    if not sentences:
        return source_text
    vectorizer = TfidfVectorizer().fit_transform([question] + sentences)
    similarities = (vectorizer[0] @ vectorizer[1:].T).toarray()[0]
    top_idx = similarities.argsort()[-n_sentences:][::-1]
    return '. '.join([sentences[i] for i in top_idx])

one_shot_dict = process_shot_files("zero_shot.csv")

# process text files
reader = SimpleDirectoryReader(
    input_dir="/content", recursive=True, required_exts=[".clean"]
)

topic_source_text = defaultdict()
for docs in reader.iter_data():
  for doc in docs:
    topic = doc.text.split("\n")[0].lower().replace("_", " ")
    topic_source_text[topic] = (doc.text.replace("\n", " "))

device = "cuda" if torch.cuda.is_available() else "cpu"

# NEW: Read questions for relevant text extraction
with open("zero_shot.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)  # skip header
    topic_questions = defaultdict(list)
    for row in reader:
        topic = row[1].lower().replace("_", " ")
        question = row[2]
        topic_questions[topic].append(question)

topic_scores = []
all_results = []

for topic, source_text in topic_source_text.items():
  if topic in one_shot_dict.keys():
    predictions = one_shot_dict[topic]
    questions = topic_questions[topic]
    references = [get_relevant_text(questions[i], source_text) for i in range(len(predictions))]

    results = bertscore.compute(predictions=predictions, references=references, lang="en", device=device)

    for i in range(len(predictions)):
      all_results.append({
          "Topic" : topic,
          "Generated_Text" : predictions[i],
          "Precision" : results["precision"][i],
          "Recall" : results["recall"][i],
          "F1" : results["f1"][i]})

    topic_scores.append({
          "Topic" : topic,
          "Total_Precision" : sum(results["precision"])/len(results["precision"]),
          "Total_Recall" : sum(results["recall"])/len(results["recall"]),
          "Total_F1" : sum(results["f1"])/len(results["f1"])})

df_scores_per_response = pd.DataFrame(all_results)
df_scores_per_topic = pd.DataFrame(topic_scores)

df_scores_per_response.to_csv("zero_shot_bert_scores_per_response.csv")
df_scores_per_topic.to_csv("zero_shot_bert_scores_per_topic.csv")
```

```python
from transformers import pipeline
import torch
import csv
from collections import defaultdict
import pandas as pd
import gc


device = 0 if torch.cuda.is_available() else -1  # 0 = first GPU, -1 = CPU
nli = pipeline("text-classification", model="cross-encoder/nli-deberta-v3-small", device=device)

def process_shot_files(path):
  topic_text_dict = defaultdict(list)
  with open(path, "r") as file:
    next(file)
    for row in csv.reader(file):
      text = row[3]
      topic = row[1].lower().replace("_", " ")
      topic_text_dict[topic].append(text)
  return topic_text_dict

one_shot_dict = process_shot_files("zero_shot.csv")

# process text files
reader = SimpleDirectoryReader(
    input_dir="/content", recursive=True, required_exts=[".clean"]
)

topic_source_text = defaultdict()
for docs in reader.iter_data():
  for doc in docs:
    topic = doc.text.split("\n")[0].lower().replace("_", " ")
    topic_source_text[topic] = (doc.text.replace("\n", " "))

topic_scores = []
all_results = []

for topic, source_text in topic_source_text.items():
  if topic in one_shot_dict.keys():
    predictions = one_shot_dict[topic]
    references = [source_text for n in range(len(predictions))]
    results = []
    for ref, hyp in zip(references, predictions):
      pair = f"{ref} </s> {hyp}"
      result = nli(pair)[0]  # [0] to extract top prediction
      results.append(result)

    for i, result in enumerate(results):
      all_results.append({
        "Topic": topic,
        "Generated_Text": predictions[i],
        "Label": result['label'],
        "Score": result['score']})

    entailment_scores = [r['score'] for r in results if r['label'] == 'ENTAILMENT']
    topic_scores.append({
    "Topic": topic,
    "Avg_Entailment_Score": sum(entailment_scores)/len(entailment_scores) if entailment_scores else 0})

    torch.cuda.empty_cache()
    gc.collect()

df_scores_per_response = pd.DataFrame(all_results)
df_scores_per_topic = pd.DataFrame(topic_scores)

df_scores_per_response.to_csv("zero_shot_deberta_per_response.csv")
df_scores_per_topic.to_csv("zero_shot_deberta_scores_pre_topic.csv")
```

```python
import os
import pandas as pd
import csv
from collections import defaultdict
from tqdm import tqdm

import torch
from transformers import pipeline
from langchain.llms import HuggingFacePipeline
from ragas.llms.base import LangchainLLMWrapper
from ragas.metrics import FaithfulnesswithHHEM
from ragas.dataset_schema import SingleTurnSample

# 1. SET UP SIMPLE HUGGINGFACE LLM (RAGAS COMPATIBLE)
# You can replace "gpt2" with a better GPU-compatible model like "tiiuae/falcon-7b-instruct" if RAM allows
hf_pipeline = pipeline("text-generation", model="gpt2", device=0 if torch.cuda.is_available() else -1)
langchain_llm = HuggingFacePipeline(pipeline=hf_pipeline)
evaluator_llm = LangchainLLMWrapper(langchain_llm)

# Initialize the faithfulness scorer
faithfulness = FaithfulnesswithHHEM(llm=evaluator_llm)

# 2. LOAD GENERATED RESPONSES CSV
df = pd.read_csv("one_shot_final.csv")  # Ensure this CSV has: Topic, Question, One_Shot_Generated_Response

# 3. LOAD SOURCE TEXTS (.clean files from /content)
topic_to_text = {}
for file in os.listdir("/content"):
    if file.endswith(".clean"):
        topic = file.replace(".clean", "").lower().replace("_", " ")
        with open(os.path.join("/content", file), "r", encoding="utf-8") as f:
            content = f.read().replace("\n", " ")
            topic_to_text[topic] = content

# 4. PREPARE & RUN EVALUATION
faithfulness_results = []
topic_agg_scores = defaultdict(list)

for idx, row in tqdm(df.iterrows(), total=len(df)):
    topic = row['Topic'].lower().replace("_", " ")
    question = row['Question']
    response = row['One_Shot_Generated_Response']
    context = topic_to_text.get(topic, "")

    if not context.strip() or "error" in response.lower():
        continue  # Skip bad rows or missing contexts

    try:
        sample = SingleTurnSample(
            user_input=question,
            response=response,
            retrieved_contexts=[context]
        )
        score = faithfulness.single_turn_score(sample)

        faithfulness_results.append({
            "Topic": topic,
            "Question": question,
            "Response": response,
            "Faithfulness": score
        })

        topic_agg_scores[topic].append(score)

    except Exception as e:
        print(f"Error at row {idx}: {e}")

# 5. SAVE RESULTS
df_results = pd.DataFrame(faithfulness_results)
df_results.to_csv("faithfulness_scores_per_response.csv", index=False)

# Aggregate scores per topic
df_topic_scores = pd.DataFrame([
    {"Topic": topic, "Avg_Faithfulness": sum(scores) / len(scores)}
    for topic, scores in topic_agg_scores.items()
])
df_topic_scores.to_csv("faithfulness_scores_per_topic.csv", index=False)

print("✅ Evaluation complete and saved.")
```
